```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('merged data.csv')
data = data.drop(columns = ['States', 'CensusId', 'IncomeErr', 'IncomePerCapErr'])
```

```python
data.info()
```

```python
data.head()
```

```python
data.info()
```

```python
#Dropping State and County data because we care about the factors that most correlate with higher incoe per capita that are not location related. We are also dropping factors that highly correlated with Income Per Capita that were not directly influencing it.
y = data['IncomePerCap']
X = data.drop(columns=['County', 'State', 'IncomePerCap', 'Poverty', 'ChildPoverty', 'Income', 'Citizen'])
X
```

We see that for variance there are a few major outlier columns. TotalPop, Men, women, and employed.

```python
X.var()
```

```python
X['TotalPop'] = np.log(X['TotalPop'])
X['Men'] = np.log(X['Men'])
X['Women'] = np.log(X['Women'])
X['Employed'] = np.log(X['Employed'])
```

```python
X.var()
```

```python
#We don't need to scale our data because we will be running regression trying to see if we can predict the Income Per Capita
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 123)
```

```python
X_train
```

```python
X_test
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
print(type(X_train))
print(type(X_test))
print(type(scaled_X_train))
print(type(scaled_X_test))
print(type(y_train))
print(type(y_test))
```

```python
scaled_X_train_df = pd.DataFrame(scaled_X_train)
scaled_X_test_df = pd.DataFrame(scaled_X_test)
```

```python
X_train.to_csv('X_train.csv', index=False)
X_test.to_csv('X_test.csv', index=False)
y_train.to_csv('y_train.csv', index=False)
y_test.to_csv('y_test.csv', index=False)
scaled_X_test_df.to_csv('scaled_X_test.csv', index=False)
scaled_X_train_df.to_csv('scaled_X_train.csv', index=False)
```
