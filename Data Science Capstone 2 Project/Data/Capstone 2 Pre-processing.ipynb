```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('merged data.csv')
data = data.drop(columns = ['States', 'CensusId', 'IncomeErr', 'IncomePerCapErr'])
```

```python
data.info()
```

```python
data.head()
```

```python
x_nominal = data[['County', 'State']]

x_nominal = OneHotEncoder(drop='first').fit_transform(x_nominal).toarray()
```

```python
type(x_nominal)
```

```python
x_nominal = pd.DataFrame(x_nominal)
x_nominal.columns = x_nominal.columns.astype(str)
```

```python
type(x_nominal)
```

```python
y = data['IncomePerCap']
X = data.drop(columns=['County', 'State', 'IncomePerCap'])
X = X.join(x_nominal)
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 123)
```

```python
sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```
